```python
import pandas as pd
df=pd.read_csv('kc-house-data.csv', sep=(','))
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df.describe()
df.info()
```

```python
df['price'].value_counts()
```

```python
df['price'].describe()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['price'])
plt.title('Распределение стоимости домов')
plt.xlabel ('Стоимость млн. USD')
plt.ylabel('Количество домов')
```

***Стоимость большинства домов лежит в диапазоне от 100 000 до 1 млн. долларов. Максимальное их количество приходится на цену около 500 000 долларов. Количество домов стоимостью более 2 млн. крайне мало.***

```python
df['sqft_living'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['sqft_living'])
plt.title('Распределение домов по размеру жилой площади')
plt.xlabel ('Жилая площадь')
plt.ylabel('Количество домов');
```

**Площадь жилая большинства домов находится в диапазоне от 250 до 4000 кв.футов. Максимальное количество домов имеет площадь 1800-2000 кв.футов. Домов площадью более 6000 футов крайне мало.**

```python
df['yr_built'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['yr_built'])
plt.title('Распределение домов по году постройки')
plt.xlabel ('Год постройки')
plt.ylabel('Количество домов');
```

## ***Количество домов по году постройки с 1940 по 2000 довольно равномерно распределено. Домов старше 1940 заметно меньше. Домов моложе 2000 заметно больше.***

```python
plt.hist(df['floors'])
plt.title('Распределение домов по количеству этажей')
plt.xlabel ('Этажи')
plt.ylabel('Количество домов');
```

```python
data1 = df['floors'].value_counts()
plt.pie(data1.values, autopct='%1.2F%%')
plt.legend(data1.index)
plt.title('Распределение домов по количеству этажей')
```

***Большинство домов - одноэтажные, далее идут двухэтажные.***

```python
data1 = df['condition'].value_counts()
plt.pie(data1.values, autopct='%1.2F%%')
plt.legend(data1.index)
plt.title('Распределение домов в зависимости от состояния')
```

# ***Большинство домов около 65% в среднем состоянии, 26% в хорошем и около 8% в отличном. Домов в плохом состоянии и состоянии ниже среднего незначительное количество - менеее 1%***

```python
sns.jointplot(x=df['yr_built'], y=df['price'], kind='reg');
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['grade'], y=df['price'], whis=1.5)
plt.title('Распределение стоимости домов в зависимости от качества конструкции здания');
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['grade'], y=df['price'], showfliers=False)
plt.title('Распределение стоимости домов в зависимости от качества конструкции здания');
```

# ***На графике очень хорошо видно, что чем выше качество конструкции здания, тем выше стоимость дома***

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['condition'], y=df['price'], showfliers=False);
plt.title('Распределение стоимости домов в зависимости от состояния');
```
